```python
import pandas as pd
import numpy as np
import cesium as cs

# Load train curves
train = pd.read_csv('data/training_set.csv')
```

```python
train.head()
```

```python
tss = train.groupby(['object_id', 'passband'])
```

```python
ts = tss.get_group((615, 2))
ts.head()
```

```python
from cesium.time_series import TimeSeries
from cesium.featurize import featurize_single_ts

demo_series = TimeSeries(
    t = ts['mjd'],
    m = ts['flux'],
    e = ts['flux_err'],
)
feat_list = [
    'amplitude',
    'flux_percentile_ratio_mid20',
    'flux_percentile_ratio_mid35',
    'flux_percentile_ratio_mid50',
    'flux_percentile_ratio_mid65',
    'flux_percentile_ratio_mid80',
    'max_slope',
    'maximum',
    'median',
    'median_absolute_deviation',
    'minimum',
    'percent_amplitude',
    'percent_beyond_1_std',
    'percent_close_to_median',
    'percent_difference_flux_percentile',
    'period_fast',
    'qso_log_chi2_qsonu',
    'qso_log_chi2nuNULL_chi2nu',
    'skew',
    'std',
    'stetson_j',
    'stetson_k',
    'weighted_average',
]
featurize_single_ts(demo_series, feat_list)
```
